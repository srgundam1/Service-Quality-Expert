Testing Accuracy of GPT Labeling on Manually Labeled Data  

```python
import pandas as pd
import numpy as np
```

```python
# Copy the filepath to the "Labeled-Texts_seed50_gpt-3.5-turbo_run3.csv"
data = pd.read_csv("Data/out/run3/Labeled-Texts_seed50_gpt-3.5-turbo_run3.csv") # import labeled data

data
```

```python
import ast

data['normalized_labels'] = data['normalized_labels'].apply(ast.literal_eval)

# Function to compare manual labels with 5D labels
def compare_labels(row):
    # Create dictionaries for quick lookup
    label_map = {'Reliability': 'man_r', 'Assurance': 'man_a', 'Responsiveness': 'man_rp', 'Tangibility': 'man_t', 'Empathy': 'man_e'}
    # Split and strip the 5D labels
    service_labels = [label.strip() for label in row['5D'].split(',')]
    # Count matches
    match_count = sum(row[label_map[label]] == 1 for label in service_labels if label in label_map)
    total_5d_labels = len(service_labels)
    total_normalized_labels = len(row['normalized_labels'])
    return match_count, total_5d_labels, total_normalized_labels

# Apply the function and expand the results into new columns
data['Matches'], data['Total_5D'], data['Total_Normalized'] = zip(*data.apply(compare_labels, axis=1))

data
```

```python
import pandas as pd
import ast

def evaluate_accuracy(df):
    df['normalized_labels'] = df['normalized_labels'].apply(ast.literal_eval)

    label_map = {
        'Reliability': 'man_r',
        'Assurance': 'man_a',
        'Responsiveness': 'man_rp',
        'Tangibility': 'man_t',
        'Empathy': 'man_e'
    }

    results = {}

    for label, column in label_map.items():
        # Check if each dimension is correctly identified in the manual labels
        df[f'{label}_correct'] = ((df['5D'].str.contains(label, na=False) | df['normalized_labels'].apply(lambda x: label.lower() in x)) == (df[column] == 1))

        # Calculate accuracy as the proportion of true values in the correctness column
        accuracy = df[f'{label}_correct'].mean()
        results[label] = accuracy

    return results


accuracy_results = evaluate_accuracy(data)

# Print the accuracy results
for dimension, acc in accuracy_results.items():
    print(f"Accuracy for {dimension}: {acc:.2%}")

data
```
