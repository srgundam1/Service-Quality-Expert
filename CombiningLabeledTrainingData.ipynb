Combining all labeled training data, removing unlabelable reviews

```python
import pandas as pd
import numpy as np
```

```python
sample1 = pd.read_csv("Data/out/training/Labeled-Training1_gpt-3.5-turbo.csv")
sample2 = pd.read_csv("Data/out/training/Labeled-Training2_gpt-3.5-turbo.csv")
sample3 = pd.read_csv("Data/out/training/Labeled-Training3_gpt-3.5-turbo.csv")
sample4 = pd.read_csv("Data/out/training/Labeled-Training4_gpt-3.5-turbo.csv")
sample5 = pd.read_csv("Data/out/training/Labeled-Training5_gpt-3.5-turbo.csv")
sample6 = pd.read_csv("Data/out/training/Labeled-Training6_gpt-3.5-turbo.csv")
sample7 = pd.read_csv("Data/out/training/Labeled-Training7_gpt-3.5-turbo.csv")

# Shuffle order while preserving Index
sample1.rename(columns={'Unnamed: 0': 'original_index'}, inplace=True)
sample2.rename(columns={'Unnamed: 0': 'original_index'}, inplace=True)
sample3.rename(columns={'Unnamed: 0': 'original_index'}, inplace=True)
sample4.rename(columns={'Unnamed: 0': 'original_index'}, inplace=True)
sample5.rename(columns={'Unnamed: 0': 'original_index'}, inplace=True)
sample6.rename(columns={'Unnamed: 0': 'original_index'}, inplace=True)
sample7.rename(columns={'Unnamed: 0': 'original_index'}, inplace=True)

vertical_concat = pd.concat([sample1, sample2, sample3, sample4, sample5, sample6, sample7], axis=0)
vertical_concat = vertical_concat.reset_index()
vertical_concat = vertical_concat.drop(columns='index')

vertical_concat
```

```python
missing = pd.DataFrame(columns=vertical_concat.columns)

for index, row in vertical_concat.iterrows():
    # Check if all specified columns are empty
    if not (row['Reliability'] or row['Tangibility'] or row['Empathy'] or row['Responsiveness'] or row['Assurance']):
        # Add a copy of the row to the missing DataFrame
        missing.loc[len(missing)] = row.copy()
        # Remove the row from vertical_concat by index
        vertical_concat = vertical_concat.drop(index)

missing
```

```python
missing.to_csv("missingdata.csv")
```

```python
sample8 = pd.read_csv("Data/out/training/Labeled-TrainingMissing8_gpt-3.5-turbo.csv")
sample8 = sample8.drop(columns="Unnamed: 0")

sample8
```

```python
final_data = pd.concat([vertical_concat, sample8], axis=0)
final_data = final_data.set_index('original_index').sort_index(ascending=True).rename_axis(None)
final_data = final_data.reset_index()
final_data = final_data.drop(columns="index")

final_data
```
