Splitting the training sample up in order to retrieve labels from ChatGPT

```python
import pandas as pd
import numpy as np
```

```python
# Upload "training_reviews.csv" into the runtime, and copy the filepath into the following
all_data = pd.read_csv("/content/training_reviews.csv")
all_data
```

```python
all_data["original_Index"] = all_data.index

# This function will take the split data and save it as a to create training samples
def split_training_data(all_data, i):
    sample = all_data.sample(n=1000)
    sampled_indices = sample.index
    remaining = all_data.drop(sampled_indices)
    print(f"{sample.count()}, {remaining.count()}, {all_data.count()}")
    sample.to_csv(f"training_sample{i}.csv", index=False, encoding="utf-8")
    return remaining

# Initial count of samples
total_samples = 6604
# Determine how many sets of 1000 can be extracted
num_full_samples = total_samples // 1000
# Loop through the full sets of 1000
for i in range(1, num_full_samples + 1):
    all_data = split_training_data(all_data, i)

# If there's any leftover, create a final smaller set
if not all_data.empty:
    all_data.to_csv(f"final_sample.csv", index=False, encoding="utf-8")
```
